```python

import pandas as pd


train_df = pd.read_csv('/content/train.csv')
train_df.head()
```

```python
eval_df = pd.read_csv('/content/val.csv')
eval_df.head()
```

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs

# Optional model configuration
model_args = ClassificationArgs(num_train_epochs=2)

# Create a TransformerModel
model = ClassificationModel('roberta', 'roberta-base',  args=model_args)
```

```python
# Train the model
model.train_model(train_df[0:1000])
```

## Evaluation

```python
def evaluate_model(model, eval_df):
    """
    Evaluates a model using the eval_model function.

    Args:
        model: The model to evaluate.
        eval_df: The dataframe containing the evaluation data.

    Returns:
        A tuple containing the evaluation result, model outputs, and wrong predictions.
    """
    result, model_outputs, wrong_predictions = model.eval_model(eval_df)
    return result, model_outputs, wrong_predictions
```

## plot

```python
import matplotlib.pyplot as plt

def plot_loss(history):
    """
    Plots the training and testing loss from a history object.

    Args:
        history: The history object containing the loss values.

    Returns:
        None
    """
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper right')
    plt.show()
```

```python

def plot_accuracy(history):
    """
    Plots the training and testing accuracy from a history object.

    Args:
        history: The history object containing the accuracy values.

    Returns:
        None
    """
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()
```
